```python
from bs4 import BeautifulSoup # html을 soup에 넣고 특정 class or data를 가져오는 library
import re
from urllib.parse import urlencode
from urllib.request import Request, urlopen
import requests
from selenium import webdriver
import os
import time
import shutil
import pandas as pd
import time
```

```python
df = pd.read_csv(".//Naver//Power_blog_url_short.txt",encoding='utf-8')
```

```python
# chorme open
driver = webdriver.Chrome('.//exe_file//chromedriver.exe')
#driver = webdriver.PhantomJS('/Users/yonggeoljo/phantomjs-2.1.1-macosx/bin/phantomjs')
driver.implicitly_wait(3)
```

## Powerblog

```python
naver_url = 'https://section.blog.naver.com/PowerBlog.nhn?searchYear='
for url in df['url']:
    driver.get(naver_url + url)
    year = url[:4]
    count = 0
    subpages = driver.find_elements_by_css_selector("#content > section > div.pagination > span > a")
    if len(subpages) != 0: 
        for subpage_num in range(0,len(subpages)):
            subpages = driver.find_elements_by_css_selector("#content > section > div.pagination > span > a")
            subpages[subpage_num].click()
            driver.implicitly_wait(3)
            title = driver.find_element_by_tag_name('h3').text
            p = re.compile('>')
            m = p.search(title)
            parent = title[:m.start()] 
            child = title[m.start()+1:]
            components = driver.find_elements_by_css_selector('#content > section > div.list_card_article > div > a.link_content')
            time.sleep(1)
            for i in range(0,len(components)):
                try:
                    components = driver.find_elements_by_css_selector('#content > section > div.list_card_article > div > a.link_content')
                    time.sleep(1)
                    components[i].click()
                    time.sleep(3)
                    driver.implicitly_wait(3)
                    window_before = driver.window_handles[0]
                    window_after = driver.window_handles[1]
                    driver.switch_to_window(window_after)
                    driver.implicitly_wait(10)
                    text = driver.current_url
                    with open('power_blog_id_remove_time.txt', 'a',encoding='utf-8') as f:
                        f.write(year + ',' + parent + ',' + child + ',' + text + '\n')
                    driver.close()
                    driver.switch_to_window(window_before)
                except:
                    print("error: ",year,parent,child,text)
    else:
        title = driver.find_element_by_tag_name('h3').text
        p = re.compile('>')
        m = p.search(title)
        parent = title[:m.start()] 
        child = title[m.start()+1:]
```

## 이달의 블로그

```python
df = pd.read_csv(".//Naver//month_blog_url_short.txt",encoding='utf-8')
```

```python
# chorme open
driver = webdriver.Chrome('.//exe_file//chromedriver.exe')
#driver = webdriver.PhantomJS('/Users/yonggeoljo/phantomjs-2.1.1-macosx/bin/phantomjs')
driver.implicitly_wait(3)
```

```python
naver_url = 'https://section.blog.naver.com/ThisMonthDirectory.nhn?year='

for url in df['url']:
    driver.get(naver_url + url)
    list_card = driver.find_elements_by_class_name('list_card_article')
    titles = driver.find_elements_by_class_name('heading_monthly_blog_categoty')

    click = driver.find_elements_by_class_name('button_more')
    count = len(click)
    for i in range(0,count):
        try:
            click = driver.find_elements_by_class_name('button_more')
            click[i].click()
        except:
            pass
    su=0
    
    for i, _ in enumerate(list_card):
        num = int(titles[i].text.split("(")[1].split(")")[0])
        child = titles[i].text.split("(")[0]
        for j in range(0,num):
            try:
                components = driver.find_elements_by_class_name('area_image')
                components[su].click()
                window_before = driver.window_handles[0]
                window_after = driver.window_handles[1]
                driver.switch_to_window(window_after)
                blog_id = driver.current_url
                year = url[:4]
                month = url[11:]
                with open('month_blog_id.txt', 'a',encoding='utf-8') as f:
                    f.write(year + '-'+ month + ',' + child + ',' + blog_id + '\n')
                driver.close()
                driver.switch_to_window(window_before)
                su+=1
            except:
                print('error: ', year + '-' + month + ',' + child + ',' + blog_id)
                su+=1
```
