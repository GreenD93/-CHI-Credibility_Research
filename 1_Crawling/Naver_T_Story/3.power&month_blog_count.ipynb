## Power blog(Count)

```python
naver_url = 'https://section.blog.naver.com/PowerBlog.nhn?searchYear='
for url in df['url']:
    driver.get(naver_url + url)
    year = url[:4]
    count = 0
    subpages = driver.find_elements_by_css_selector("#content > section > div.pagination > span > a")
    if len(subpages) != 0: 
        for subpage_num in range(0,len(subpages)):
            subpages = driver.find_elements_by_css_selector("#content > section > div.pagination > span > a")
            subpages[subpage_num].click()
            time.sleep(3)
            title = driver.find_element_by_tag_name('h3').text
            components = driver.find_elements_by_class_name("blog_info")
            count += len(components)
            p = re.compile('>')
            m = p.search(title)
            parent = title[:m.start()] 
            child = title[m.start()+1:]
    else:
        title = driver.find_element_by_tag_name('h3').text
        count = 0
        p = re.compile('>')
        m = p.search(title)
        parent = title[:m.start()] 
        child = title[m.start()+1:]
        
    with open('power_blog_count_real.txt', 'a',encoding='utf-8') as f:
        f.write(year + ',' + parent + ',' + child + ',' + str(count) + '\n')
    print(year,parent,child,str(count))
```

```python
df = pd.read_csv("month_blog_url_short.txt",encoding='utf-8')
```

```python
naver_url = 'https://section.blog.naver.com/ThisMonthDirectory.nhn?year='

for url in df['url']:
    driver.get(naver_url + url)
    e = driver.find_elements_by_class_name('heading_monthly_blog_categoty')
    for i in e:
        year = url[:4]
        month = url[11:]
        child = i.text.split("(")[0]
        count = i.text.split("(")[1].split(")")[0]
        with open('month_blog_count_real.txt', 'a',encoding='utf-8') as f:
            f.write(year+'-'+ month + ',' + ' ' + ',' + child + ',' + str(count) + '\n')
            print(year+'-'+ month + ',' + ' ' + ',' + child + ',' + str(count) + '\n')
        time.sleep(3)
```

```python
power_blog = pd.read_csv("power_blog_count_real.txt",sep=',',names=['year','parent','child','count'])
month_blog = pd.read_csv("month_blog_count_real.txt",sep=',',names=['year','parent','child','count'])
```

```python
total_blog = pd.concat([power_blog,month_blog],ignore_index=True)
```

```python
total_blog.to_csv('Naver_total_blog.csv',encoding='euc-kr',index=None)
```
