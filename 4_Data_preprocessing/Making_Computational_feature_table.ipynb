## Avg_sent_score, FR_SENT_POS, FR_SENT_NEG, Subjectivity, Polarity

```python
sentiment = pd.read_csv('polarity.csv')
```

```python
sentiment[['ngram','max.value']]
word_list = sentiment['ngram'].tolist()
label = sentiment['max.value'].tolist()
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
def sentimental_analysis(text):
    if text == '':
        return 0.000000001, 0.000000001, 0.000000001, 0.000000001, 0.000000001
    else:
        pos = 0
        neg = 0
        text = text.split(' ')
        n = len(text)
        for i in text:
            pre = kkma.pos(i)
            test = ';'.join(['/'.join(i) for i in pre])
            if test in word_list:
                if label[word_list.index(test)] == 'POS':
                    pos +=1
                elif label[word_list.index(test)] == 'NEG':
                    neg +=1
        try:
            pos_ratio = pos/n
        except:
            pos_ratio = 0.000000001
        try:
            neg_ratio = neg/n
        except:
            neg_ratio = 0.000000001
        try:
            subjectivity = (neg+pos)/n
        except:
            subjectivity = 0.000000001
        try:
            polarity = (neg-pos)/(neg+pos)
        except:
            polarity = 0.000000001
        try:
            senti_diffs_per_ref = (pos-neg)/n
        except:
            senti_diffs_per_ref = 0.000000001
            
        return pos_ratio, neg_ratio, subjectivity, polarity, senti_diffs_per_ref
```

```python
Text = Content_df['Text'].tolist()
```

```python
pos_ratio = []
neg_ratio = []
subjectivity = []
polarity = []
senti_diffs_per_ref = []
for i in Text:
    sentiment_result = sentimental_analysis(i)
    pos_ratio.append(sentiment_result[0])
    neg_ratio.append(sentiment_result[1])
    subjectivity.append(sentiment_result[2])
    polarity.append(sentiment_result[3])
    senti_diffs_per_ref.append(sentiment_result[4])
```

```python
Content_df['pos_ratio'] = pos_ratio
Content_df['neg_ratio'] = neg_ratio
Content_df['subjectivity'] = subjectivity
Content_df['polarity'] = polarity
Content_df['senti_diffs_per_ref'] = senti_diffs_per_ref
```

## FR_PRON_FIRST, Second personal pronounce sentence rate

```python
Text = Text[15000:]
```

```python
Text[1062] = ''
```

```python
from tqdm import tqdm
```

```python
first_person = ['나/NP','저/NP','내/NP','제/NP','저희/NP','우리/NP']
second_person = ['너/NP','자네/NP','당신/NP','그대/NP','그쪽/NP','너희/NP','자기/NP']
```

```python
def check_First_second(Text):
    if Text == '':
        return 0, 0
    else:
        First = 0
        Second = 0
        text = kkma.pos(Text)
        for i in text:
            temp = "/".join(i)
            if temp in first_person:
                First += len(temp.split('/')[0])
            if temp in second_person:
                Second += len(temp.split('/')[0])
        return First, Second
```

```python
First_list = []
Second_list = []
for i in tqdm(Text):
    sentiment_result = check_First_second(i)
    First_list.append(sentiment_result[0])
    Second_list.append(sentiment_result[1])
```
